```python
import seaborn as sns
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
sns.set()
```

```python
data_sample = pd.read_csv('DataSample__201709211530.csv')
data_workout = pd.read_csv('DataWorkoutRun__201709211532.csv')
```

```python
# 'sourceId', 'sourceName', 'track', 'originalData' columns are all null...
data_workout[['sourceId', 'sourceName', 'track', 'originalData']].isnull().sum() == len(data_workout)
```

```python
# 'source' column has only one non-null value...
data_workout['source'].unique()
```

```python
# ...so let's get rid of them
data_workout = data_workout.drop(['source', 'sourceId', 'sourceName', 'track', 'originalData'], axis=1)
# and let's get rid of the 'createdAt' and 'updatedAt' fields because we don't need them
data_workout = data_workout.drop(['createdAt', 'updatedAt'], axis=1)
```

```python
# convert the run data into floats
to_float = lambda x: float(x.replace(',',''))
data_workout[['runDuration', 'runDistance']] = data_workout[['runDuration', 'runDistance']].applymap(to_float)
```

```python
data_workout[['runDuration', 'runDistance']].describe()
```

```python
# 43 entries (out of 213) have a 'runDistance' of 0
zero_run_dist = data_workout[data_workout['runDistance'] == 0]
len(zero_run_dist), str(round(len(zero_run_dist)/len(data_workout)*100, 1)) + '% of all entries'
```

```python
# but every entry still has a non-zero 'runDuration'
len(data_workout[data_workout['runDuration'] == 0])
```

```python
# converting 'startDate' column to Timestamp objects
data_workout['startDate'] = pd.to_datetime(data_workout['startDate'], format='%Y-%m-%d %X')
```

```python
from datetime import datetime
to_day = lambda x: datetime.strftime(x, '%Y-%m-%d')
grouped_by_day = data_workout.groupby(data_workout['startDate'].map(to_day))
grouped_by_day.count()
```

```python
grouped_by_day.sum()
```

```python
# some of the entries with 0 'runDistance' get combined with other entries from the same day
# that have a nonzero 'runDistance', but many don't
daily_totals = grouped_by_day.sum()
len(daily_totals[daily_totals['runDistance'] == 0])
```

```python
# on this day, the first two entries are almost identical. it appears that they might've ended
# around the same time, but the first was started one second earlier
data_workout.iloc[11:14].drop(['startLocation', 'stopLocation'], axis=1)
```

```python
# maybe if we group it based on ten minute intervals (this is rough b/c it won't work if the 
# two start times are in different ten minute intervals even if they are a second apart; also
# it's going to put some runs into a ':60' group) the duplicates will be removed
to_ten_minutes = lambda x: datetime.strftime(x, '%Y-%m-%d %H:') + str(round(int(datetime.strftime(x, '%M')),-1))
grouped_by_tens = data_workout.groupby(data_workout['startDate'].map(to_ten_minutes))
grouped_by_tens.count()
```

```python
# this should remove (many of) those duplicates
grouped_by_tens.max()
```

```python
# just checking to make sure that the groups don't contain different workouts... turns out
# one does
sum(grouped_by_tens['sessionTitle'].nunique()) - len(grouped_by_tens)
```

```python
# not sure what to make of this
grouped_by_tens.get_group(grouped_by_tens['sessionTitle'].nunique().idxmax())
```

```python
# let's try this again...
to_ten_min_type = lambda row: to_ten_minutes(row['startDate']) + ' ' + row['sessionTitle']
grouped_by_session = data_workout.groupby(data_workout[['startDate','sessionTitle']].apply(to_ten_min_type, axis=1))
data_workout_no_d = grouped_by_session.max()
data_workout_no_d
```

```python
# 'nan' might appear like it should be grouped into 'Run' (?)
data_workout_no_d['trackTitle'].unique()
```

```python
tracks = data_workout_no_d['trackTitle'].unique()[:-1]
tracks
```

```python
# (thanks John)
def graph(i):    
    graph = data_workout_no_d[data_workout_no_d['trackTitle'] == i][['startDate','runDuration', 'runDistance']].set_index('startDate').plot()
    graph.set_ylabel(i)

for track in tracks:
    graph(track)
```
